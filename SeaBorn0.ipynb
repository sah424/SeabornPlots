```python
'''
The ways of styling themes are as follows :
1.white
2.dark
3.whitegrid
4.darkgrid
5.ticks
Usages -> sns.set_style('whitegrid')


set_context() allows us to override default parameters. This affects things like the size of the labels,lines and other elements of the plot,
but not the overall style -> set_style()
The contexts are as follows (According to size):
1.poster - Suitable for large displays and posters(Very large elements)
2.talk - Good for presentation slides where readability matters(Larger text and lines)
3.notebook - Good for working inside Jupyter notebooks(default settings)(Medium size)
4.paper - Ideal for printing in academic papers(Smallest elements)
Usages -> sns.set_context('paper')

hue -> It makes billboard of the identites, it can devide the data based on the identity/parameter passed
       Category you want to separete your datapoint with

Palette -> It will change the colour of the point accordingly
Palette Type  ----   Tags
Qualitative   ----   'set1','set2','Paired','Pastel1','Pastel2','Accent','Dark2','Colorblind'
Squential     ----   'Blues','BuGn','BuPu','GnBu','Greens','Greys','Oranges','OrRd','Purples','RdPu','Reds','YlGn','YlGnBu','YlOrBr','YlOrRd'
Diverging     ----   'BrBG','PiYG','PRGn','PuOr','RdBu','RdGy','coolwarm','bwr','seismic'
'''
```

```python
# scatterplot
sns.set_style('darkgrid')
sns.set_context('paper')
sns.scatterplot(data = penguins, x = "flipper_length_mm", y = "body_mass_g",  hue = "island", style = "sex",palette = 'Dark2' )
plt.savefig('scatter.png', dpi = 300, bbox_inches = 'tight')
# This is for removing spines from all 4 sides
# sns.despine(left = True)
```

```python
# stripplot
sns.stripplot(data = penguins, x ="species", y ="body_mass_g", hue = "island",dodge = True, jitter = True)
plt.savefig('stripplot.png', dpi = 300, bbox_inches = 'tight')
# dodge and jitter both helps to read plot better
# dodge -> Makes Adile into 3 different identity
```

```python
# Swarmplot -> Gave swarm shape data
sns.swarmplot(data = penguins, x ="species", y ="body_mass_g", hue = "island")
plt.savefig('swarmplot.png', dpi = 300, bbox_inches = 'tight')
```

```python
# histogram
sns.histplot(data = penguins, x = "body_mass_g", hue = "sex")
plt.savefig('histogram.png', dpi = 300, bbox_inches = 'tight')
```

```python
# regression plot
sns.regplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm")
plt.savefig('regression.png', dpi = 300, bbox_inches = 'tight')
```

```python
# Line Plot
sns.set_style("whitegrid")
sns.lineplot(data = penguins, x = "body_mass_g", y = "flipper_length_mm", hue="island", style = "sex")
plt.savefig('lineplot.png', dpi = 300, bbox_inches = 'tight')
```

```python
# Bar plot
sns.barplot(data = penguins, x = "species", y = "body_mass_g", hue = "sex", palette = ["skyblue","pink"])
plt.savefig('barplot.png', dpi = 300, bbox_inches = 'tight')
```

```python
# JointPlot ->Histogram + Scatter plot(By Default)
sns.jointplot(data=penguins, x="body_mass_g", y="flipper_length_mm")
plt.savefig('Histogram_Scatter_JP.png', dpi = 300, bbox_inches = 'tight')
```

```python
# Joint Plot = Scatter Plot + Density Plot(because here we decsribed hue)
sns.jointplot(data=penguins, x="body_mass_g", y="flipper_length_mm", hue ="sex")
plt.savefig('Scatter_Density_JP.png', dpi = 300, bbox_inches = 'tight')
```

```python
# Joint Plot = Scatter Plot + KDE Plot
sns.jointplot(data=penguins, x="body_mass_g", y="flipper_length_mm", kind = "kde", hue ="sex")
plt.savefig('Scatter_KDE_JP.png', dpi = 300, bbox_inches = 'tight')
# kind only can be -
# 'scatter'
# 'hist'(hue and kind both desn't goes in hist)
# 'hex'(hue and kind both desn't goes in hex)
# 'kde' (Kernal Density Explanation)
# 'reg'(hue and kind both desn't goes in reg)
# 'resid'(hue and kind both desn't goes in resid)
```

```python
# Countplot -> To count Data
sns.countplot(data = penguins, x = "species", hue = "sex")
plt.savefig('Countplot.png', dpi = 300, bbox_inches = 'tight')
```

```python
# Box plot -> Helps to identify the different groups
# - -> Top Value
# Quartiles
# Upper Quartile -> 100%
# Median -> 75%
# Lower -> 25%
# - -> bottom Value
# o -> Goes for extremly high or low value

sns.boxplot(data = penguins, x = "species", y = "body_mass_g", hue = "island")
plt.savefig('Boxplot.png', dpi = 300, bbox_inches = 'tight')
```

```python
# Violin Plot
'''
Width Of the Violin - Density of the data at differnt values
Middle Line - Median
Thick Bar - Interquartile range (if inner = box)
'''
sns.violinplot(data = penguins, x = "species", y = "body_mass_g", hue = "sex", split = True)
plt.savefig('Violinplot.png', dpi = 300, bbox_inches = 'tight')
'''
In Seaborn, the split parameter (primarily in violinplot()) divides the,violin plots in half along the categorical axis,
allowing direct, "face-to-face" comparison of two different distributions (usually defined by a hue` variable) within the same category.
It reduces visual clutter by merging two violins into one, with each side representing a different subset of data
'''
```

```python
# sns.jointplot(data = penguins, x = "species", y = "body_mass_g", hue = "sex", kind = "swarm")
sns.violinplot(data = penguins, x="species", y = "body_mass_g")
sns.swarmplot(data = penguins, x="species", y = "body_mass_g", hue ="sex", color = "black", size = 3)
plt.savefig('Violin_Swarm_merge.png', dpi = 300, bbox_inches = 'tight')
# It will join each other
# But hue mass has to pass threw swarmplot
```

```python
# KDE - Kernel Density Estimate plot
# For estimating the probability density function of continuous or non-parametric data
# Or in simple terms - a smooth curve that shows the ditribution of data
# Like a smoothed-out version of a histrogram
# It helps you see where the data is concentrated(peaks) and how it spreads

sns.kdeplot(data = penguins, x = "body_mass_g", hue = "species", fill = True)
plt.savefig('KDE.png', dpi = 300, bbox_inches = 'tight')
```

```python
columns = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
penguins[columns].corr()
```

```python
  # Heatmap - A 2D color coded matrix that shows relationships or pattern bw 2 variables
  # It's commonly used to
  #   - Show co-relation bw features
  #   - Visualize confusion matrices
  #   - Highlight missing data -> sns.heatmap(df.isnull())
sns.heatmap(data = penguins[columns].corr(), annot = True, cmap ='Blues', vmin = -0.2, linewidth = '2', linecolor ='black')
plt.savefig('Heatmap.png', dpi = 300, bbox_inches = 'tight')
```

```python
# Rug Plot -> Makes small ticks/line along the axis which user defined, reffers differnt data column or feature of the column
sns.rugplot(data = penguins, y = "body_mass_g", hue = "species", palette = "Set2", height = 0.5)
plt.savefig('Rugplot.png', dpi = 300, bbox_inches = 'tight')
```

```python
# pair plot -> Grid of plots that shows the relation bw all the numerical variables in a df
# It works only with numerical columns,don't need to pass numerical clm explicitly
sns.pairplot(data = penguins, hue = "sex", palette = "Set2", diag_kind = "kde")
plt.savefig('Pairplot.png', dpi = 300, bbox_inches = 'tight')
```

```python
# pair grid -> subplot grid for plotting pairwise relationship in a dataset.
graph = sns.PairGrid(data = penguins, hue = "sex", palette = "Set2")
graph.map_upper(sns.scatterplot) #upper triangle
graph.map_lower(sns.kdeplot) #lower triangle
graph.map_diag(sns.histplot) #diagonal

graph.add_legend() #to show which color for which sex

plt.savefig('Pairgrid.png', dpi = 300, bbox_inches = 'tight')
```
